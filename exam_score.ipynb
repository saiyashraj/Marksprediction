```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.columns
```

```python
features=['study_hours_per_day','attendance_percentage','mental_health_rating', 'sleep_hours','part_time_job']
```

```python
targets='exam_score'
```

```python
df_model = df[['study_hours_per_day', 'attendance_percentage', 'mental_health_rating', 'sleep_hours', 'part_time_job','exam_score']]
```

```python
df_model
```

```python
le=LabelEncoder()
import warnings
warnings.filterwarnings('ignore')
```

```python
df_model['part_time_job']=le.fit_transform(df_model['part_time_job'])
```

```python
df_model
```

```python
X=df_model[features]
```

```python
Y=df_model[targets]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
models = {
    'Linear_regression': {
        'model': LinearRegression(),
        'params': {}
    },
    'DecisionTree': {
        'model': DecisionTreeRegressor(),
        'params': {
            'max_depth': [3, 5, 10],
            'min_samples_split': [2, 5]
        }
    },
    'RandomForest': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [50, 100],
            'max_depth': [100]
        }
    }
}
```

```python
best_models=[]
```

```python
for name,config in models.items():
    print(f"Traning{name}")

    grid=GridSearchCV(config['model'],config['params'],cv = 5,scoring ='neg_mean_squared_error')
    grid.fit(X_train,Y_train)

    Y_pred = grid.predict(X_test)
    rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
    r2=r2_score(Y_test,Y_pred)

    best_models.append({
    "model":name,
    "best_param":grid.best_params_,
    "rmse":rmse,
    "r2":r2})
```

```python
best_models
```

```python
results=pd.DataFrame(best_models)
```

```python
results.sort_values(by='rmse')
```

```python
import joblib
best_row=results.sort_values(by='rmse').iloc[0]
```

```python
best_row
```

```python
best_model_name=best_row['model']
```

```python
best_model_name
```

```python
best_model_config = models[best_model_name]
```

```python
best_model_config
```

```python
final_model=best_model_config['model']
```

```python
final_model.fit(X,Y)
```
